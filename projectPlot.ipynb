```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


pd.options.display.max_rows = 20
pd.options.display.max_columns = 100
```

```python
alsDF = pd.read_csv('.\\alsRst.csv')
alsDF.head()
```

```python
clDF = pd.read_csv('.\\clRst.csv')
clDF.head()
```

```python
singleDF = pd.read_csv('.\\singleMachineTestTime.csv')
singleDF.head()
```

```python
bestALSDF = pd.read_csv('.\\bestResultForALS.csv')
bestALSDF.head()
```

```python
dfRank20RMSE = alsDF[alsDF.rankOfALS == 20]
dfRank20RMSE
```

```python
dfRank50RMSE = alsDF[alsDF.rankOfALS == 50]
dfRank50RMSE
```

```python
dfRank100RMSE = alsDF[alsDF.rankOfALS == 100]
dfRank100RMSE
```

```python
dfRank200RMSE = alsDF[alsDF.rankOfALS == 200]
dfRank200RMSE
```

```python
dfRank300RMSE = alsDF[alsDF.rankOfALS == 300]
dfRank300RMSE
```

```python
plt.figure(figsize=(18,9))
x = np.arange(11)
widthValue = 0.15
plt.bar(x-widthValue*2, dfRank20RMSE['DataSize'], width = widthValue, label='DataSize', facecolor = 'IndianRed')
plt.xticks(x, dfRank20RMSE['cityName'])
plt.legend(bbox_to_anchor=(1.1, 1))
plt.title("DataSize of different cities")
plt.xlabel("City Names") 
plt.ylabel("DataSize(rows)")
plt.show()
```

```python
plt.figure(figsize=(18,9))
x = np.arange(5)
widthValue = 0.15
plt.bar(x-widthValue*2, dfRank20RMSE['RMSE'], width = widthValue, label='RMSE for rank 20')
plt.bar(x-widthValue, dfRank50RMSE['RMSE'], width = widthValue, label='RMSE for rank 50')
plt.bar(x, dfRank100RMSE['RMSE'], width = widthValue, label='RMSE for rank 100')
plt.bar(x+widthValue*1, dfRank200RMSE['RMSE'], width = widthValue, label='RMSE for rank 200')
plt.bar(x+widthValue*2, dfRank300RMSE['RMSE'], width = widthValue, label='RMSE for rank 300')
plt.xticks(x, dfRank20RMSE['cityName'])
plt.legend(bbox_to_anchor=(1, 1))
plt.title("ALS RMSEs of different rank values")
plt.xlabel("City Names") 
plt.ylabel("RMSE")
plt.show()
```

```python
plt.figure(figsize=(18,9))
x = np.arange(11)
widthValue = 0.15
plt.bar(x-widthValue*2, dfRank20RMSE['trainTime'], width = widthValue, label='trainTime for rank 20')
plt.bar(x-widthValue, dfRank50RMSE['trainTime'], width = widthValue, label='trainTime for rank 50')
plt.bar(x, dfRank100RMSE['trainTime'], width = widthValue, label='trainTime for rank 100')
plt.bar(x+widthValue*1, dfRank200RMSE['trainTime'], width = widthValue, label='trainTime for rank 200')
plt.bar(x+widthValue*2, dfRank300RMSE['trainTime'], width = widthValue, label='trainTime for rank 300')
plt.xticks(x, dfRank20RMSE['cityName'])
plt.legend(bbox_to_anchor=(1, 1))
plt.title("ALS training time of different rank values")
plt.xlabel("City Names") 
plt.ylabel("trainTime(s)")
plt.show()
```

```python
plt.figure(figsize=(15,9))
x = np.arange(5)
widthValue = 0.15
plt.bar(x-widthValue, clDF[clDF.neighborNumber == 15]['RMSE'], width = widthValue, 
        label='RMSE for neighborNumber 15', facecolor = 'IndianRed')
plt.bar(x, clDF[clDF.neighborNumber == 20]['RMSE'], width = widthValue, 
        label='RMSE for neighborNumber 20', facecolor = 'yellowgreen')
plt.bar(x+widthValue, clDF[clDF.neighborNumber == 30]['RMSE'], width = widthValue, 
        label='RMSE for neighborNumber 30', facecolor = 'SkyBlue')
plt.xticks(x, clDF[clDF.neighborNumber == 15]['cityName'])
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Item-Based CF RMSEs of different neighbor numbers")
plt.xlabel("City Names") 
plt.ylabel("RMSE")
plt.show()
```

```python
x = np.arange(5)
widthValue = 0.15
plt.bar(x-widthValue, clDF[clDF.neighborNumber == 15]['trainTime'], width = widthValue, 
        label='Train Time for neighborNumber 15')
plt.bar(x, clDF[clDF.neighborNumber == 20]['trainTime'], width = widthValue, 
        label='Train Time for neighborNumber 20')
plt.bar(x+widthValue, clDF[clDF.neighborNumber == 30]['trainTime'], width = widthValue, 
        label='Train Time for neighborNumber 30')
plt.xticks(x, clDF[clDF.neighborNumber == 15]['cityName'])
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Item-Based CF Train Time of different neighbor numbers")
plt.xlabel("City Names") 
plt.ylabel("Train Time(s)")
plt.show()

plt.bar(x-widthValue, clDF[clDF.neighborNumber == 15]['testTime'], width = widthValue, 
        label='Train Time for neighborNumber 15')
plt.bar(x, clDF[clDF.neighborNumber == 20]['testTime'], width = widthValue, 
        label='Train Time for neighborNumber 20')
plt.bar(x+widthValue, clDF[clDF.neighborNumber == 30]['testTime'], width = widthValue, 
        label='Train Time for neighborNumber 30')
plt.xticks(x, clDF[clDF.neighborNumber == 15]['cityName'])
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Item-Based CF Test Time of different neighbor numbers")
plt.xlabel("City Names") 
plt.ylabel("Train Time(s)")
plt.show()
```
